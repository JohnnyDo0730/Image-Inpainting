```python
import os
import re

def generate_image_mask_list(image_folder, mask_folder, result_path):
    # Helper function to extract numeric and logical ordering from filenames
    def extract_numeric_key(filename):
        matches = re.findall(r'\d+', filename)
        return [int(match) for match in matches] if matches else [float('inf')]

    # Sort files based on numeric parts of filenames
    image_files = sorted(os.listdir(image_folder), key=extract_numeric_key)
    mask_files = sorted(os.listdir(mask_folder), key=extract_numeric_key)

    # Ensure sorted file lists have the same length
    if len(image_files) != len(mask_files):
        raise ValueError("Image and mask folder do not contain the same number of files.")

    # Combine corresponding paths into the desired format
    result_lines = []
    for img_file, mask_file in zip(image_files, mask_files):
        img_path = os.path.join(image_folder, img_file)
        mask_path = os.path.join(mask_folder, mask_file)
        result_lines.append(f"{img_path}\t{mask_path}\n")

    # Write results to the specified file
    with open(result_path, 'w') as result_file:
        result_file.write("".join(result_lines))

    # Return the result path
    return result_path
```
